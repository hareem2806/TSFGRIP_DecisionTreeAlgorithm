# **Label Encoding**

```python
df['Species']
```

***As we know that our target variable Species is catagorical and of object datatype so we need to perform label encoding***

```python
#using label encoding for converting categories(character dtype) to numerical representation for making models models work on numerics only
#encoding categorical data
labelEncode=LabelEncoder()
df['Species']=labelEncode.fit_transform(df['Species'])

```

```python
df['Species']
```

0 = Iris-setosa

# **Splitting into Testing and Training Data**

```python
X=df.drop(columns=['Species'])
Y=df['Species']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.20)
```

```python
X_test
```

```python
Y_test
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(criterion = 'entropy', max_depth = 4)
```

```python
clf.fit(X_train, Y_train)
```

```python
print("Accuracy: ",clf.score(X_test, Y_test)*100)
```

```python
from sklearn.tree import plot_tree
from sklearn import tree
fn = ['SepalLengthCm',	'SepalWidthCm',	'PetalLengthCm'	,'PetalWidthCm']

# Correctly get class labels as strings:
class_labels = df['Species'].unique().astype(str).tolist()
clf.fit(X_train, Y_train)
# Create figure and axes:
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(10, 6), dpi=300)

# Plot the decision tree with correct arguments:
plot_tree(clf, feature_names=fn, class_names=class_labels, filled=True, ax=axes)
plt.show()  # Display the plot
```

```python
# Get predictions for all test instances
from sklearn import metrics
y_pred3 = clf.predict(X_test)

print(metrics.classification_report(Y_test, y_pred3))
```

```python
#PRINTING THE CONFUSION MATRIX

print("Confusion Matrix\n\n")
print(metrics.confusion_matrix(Y_test, y_pred3))
```

```python
test_data = {
    'SepalLengthCm': [5.1, 6.2, 4.7, 5.5],
    'SepalWidthCm': [3.5, 2.9, 3.2, 4.0],
    'PetalLengthCm': [1.4, 4.3, 1.6, 1.5],
    'PetalWidthCm': [0.2, 1.3, 0.2, 0.2]
}

# Create a DataFrame for the test data
test_df = pd.DataFrame(test_data)

# Now you can use this test data to make predictions
predictions = clf.predict(test_df)

# Print the predictions
print("Predictions:")
for prediction in predictions:
    print(prediction)
```
